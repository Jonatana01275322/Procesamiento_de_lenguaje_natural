*   **Pregunta 1.** 

Busca y elimina todos los saltos de línea '\n' que se encuentran al final de cada comentario. 

Una vez finalizado, imprime los primeros 10 comentarios del resultado obtenido.


```python
# Usamos una lista de comprensión para aplicar re.sub a cada comentario (c) en la lista docs.
# El patrón r'\n$' busca el salto de línea específicamente al final de la cadena.

com_clean = [re.sub(r'\n$', '', c) for c in docs]
```

```python
# Imprimimos los primeros 10 elementos de la lista ya limpia para verificar el cambio.
print(f"Comentarios sin salto de línea: {com_clean[:10]}")

# Mostramos el conteo total de elementos para asegurar que seguimos teniendo los 1000 originales.
print(f"Total de comentarios sin el salto de línea: {len(com_clean)}")
```

*   **Pregunta 2.**  

Busca e imprime todas las palabras que terminan con dos o más signos de admiración seguidos, por ejemplo "!!!". 

Debes imprimir tanto la palabra como la totalidad de signos de admiración que le siguen. 

Indica cuántos resultados obtuviste.



```python
# Unimos todos los comentarios en un solo string y buscamos palabras (\w+) 
# que terminen con 2 o más signos de admiración (!{2,}) de forma consecutiva.

sign_adm = re.findall(r'\w+!{2,}', " ".join(docs))
```

```python
# Imprimimos el conteo total de palabras que cumplen con el patrón de énfasis.
print(f"Total de palabras encontradas: {len(sign_adm)}")

# Mostramos las palabras encontradas.
print(f"Palabras encontradas: {sign_adm}")
```

*   **Pregunta 3.**  

Busca e imprime todas las palabras que están escritas totalmente en mayúsculas. Cada coincidencia debe ser una sola palabra.

Indica cuántas palabras encontraste.



```python
# Utilizamos findall para buscar todas las ocurrencias que coincidan con el patrón.
# \b marca el límite de una palabra, y [A-ZÁÉÍÓÚÑ]+ busca uno o más caracteres en mayúscula, incluyendo acentos.

mayus_words = re.findall(r'\b[A-ZÁÉÍÓÚÑ]+\b', " ".join(docs))
```

```python
# Mostramos la cantidad total de palabras encontradas que cumplen con ser totalmente mayúsculas.
print(f"Total de palabras encontradas: {len(mayus_words)}")

# Imprimimos todos los elementos de la lista para verificar el tipo de palabras capturadas.
print(f"Muestra de palabras en mayúsculas: {mayus_words}")
```

*   **Pregunta 4.**  

Busca e imprime los comentarios en donde todos los caracteres alfabéticos (letras) están en mayúsculas. 

Cada coincidencia encontrada debe ser todo el comentario/enunciado.

Indica cuántos resultados obtuviste.


```python
# Filtramos comentarios que contienen al menos una letra mayúscula y ninguna letra minúscula.
full_mayus = [c for c in com_clean if re.search(r'[A-Z]', c) and not re.search(r'[a-z]', c)]
```

```python
# Mostramos una muestra de los primeros 5 comentarios que cumplen con estar escritos totalmente en mayúsculas.
print(f"Muestra de comentarios en mayúsculas: {full_mayus[:5]}")

# Imprimimos el total de comentarios encontrados que coinciden con el patrón de "solo mayúsculas".
print(f"Total de comentarios detectados: {len(full_mayus)}")
```

*   **Pregunta 5.**  

Busca e imprime todas las palabras que tengan una vocal acentuada, del tipo á, é, í, ó, ú. 

Indica cuántos resultados obtuviste.

```python
# Utilizamos re.findall para buscar palabras (\w*) que contengan al menos una vocal acentuada [áéíóúÁÉÍÓÚ] en cualquier posición.
vocal_words = re.findall(r'\b\w*[áéíóúÁÉÍÓÚ]\w*\b', " ".join(com_clean))
```

```python
# Mostramos la lista de palabras encontradas para verificar.
print(f"Palabras con vocales acentuadas: {vocal_words}")

# Imprimimos el total de coincidencias.
print(f"Total de resultados obtenidos: {len(vocal_words)}")
```

*   **Pregunta 6.**  

Busca e imprime todas las cantidades numéricas monetarias, enteras o con decimales, que inician con el símbolo $\$$. 

Indica cuántos resultados obtuviste.

```python
# Buscamos en todo el texto unido (join) todas las coincidencias que empiecen con el símbolo $ seguido de dígitos y posibles separadores decimales.
cant_dinero = re.findall(r'\$\d+(?:[.,]\d+)*', " ".join(docs))
```

```python
# Imprimimos el total de menciones monetarias encontradas para conocer la frecuencia de precios en el corpus.
print(f"Total de cantidades monetarias encontradas: {len(cant_dinero)}")

print(f"Cantidades monetarias encontradas: {cant_dinero}")
```

*   **Pregunta 7.**  

Busca e imprime todas las palabras que sean variantes de la palabra "love", sin importar si incluyen mayúsculas o minúsculas, o la manera en que esté conjugada o alguna otra variación que se haga con dicha palabra.

Indica cuántos resultados obtuviste.

```python
# Utilizamos re.findall con la bandera re.IGNORECASE para encontrar todas las palabras que comiencen con la raíz 'lov'.
# El patrón \blov\w*\b asegura que capturemos la palabra completa (como 'loved' o 'loving') independientemente de su terminación.
love_variants = re.findall(r'\blov\w*\b', " ".join(docs), re.IGNORECASE)
```

```python
# Imprimimos la lista completa de las variantes encontradas para analizar las distintas formas de expresar afecto en las reseñas.
print(f"Variantes de 'love' encontradas: {love_variants}")

# Mostramos el conteo total de variantes encontradas.
print(f"Total de resultados obtenidos: {len(love_variants)}")
```

*   **Pregunta 8.**  

Busca e imprime todas las palabras, variantes de "so" y "good", que tengan dos o más "o" en "so" y 3 o más "o" en good. 

Indica cuántas encontraste.


```python
# Buscamos variantes de "so" con 2 o más letras "o" y de "good" con 3 o más letras "o".
so_good_variants = re.findall(r'\bso{2,}\b|\bgo{3,}d\b', " ".join(docs), re.IGNORECASE)
```

```python
# Imprimimos la lista de las exageraciones encontradas para ver cómo los usuarios enfatizan sus reseñas.
print(f"Exageraciones encontradas: {so_good_variants}")

# Mostramos el total de coincidencias encontradas.
print(f"Total: {len(so_good_variants)}")
```

*   **Pregunta 9.**  

Busca e imprime todas las palabras que tengan una longitud mayor estrictamente a 10 caracteres alfabéticos.

No se consideran los signos de puntuación o caracteres especiales en la longitud de estas cadenas, solo caracteres alfabéticos en mayúsculas o minúsculas.

Indica la cantidad de palabras encontradas.


```python
palabras_largas = [p for p in re.findall(r'\b[A-Za-z]+\b', " ".join(docs)) if len(p) > 10 ]
```

```python
# Imprimimos la lista completa de palabras largas encontradas.
print(f"Palabras de más de 10 letras: {palabras_largas}")

# Mostramos el conteo total de palabras encontradas.
print(f"Total encontrado: {len(palabras_largas)}")
```

*   **Pregunta 10.**  

Busca e imprime todas las palabras que inician con una letra mayúscula y terminan con una minúscula, pero que además no sea la primera palabra del comentario/string. 

Indica la cantidad de resultados obtenidos.

```python
# Busca palabras por comentario, excluye la primera palabra de cada uno con [1:],
# y conserva solo las que inician con mayúscula y terminan con minúscula.
palabras_p10 = [
    palabra
    for comentario in docs
    for palabra in re.findall(r'\b[A-Za-zÁÉÍÓÚÑáéíóúñ]+\b', comentario)[1:]
    if re.fullmatch(r'[A-ZÁÉÍÓÚÑ][A-Za-zÁÉÍÓÚÑáéíóúñ]*[a-záéíóúñ]', palabra)
]
```

```python
# Imprimimos la lista de palabras encontradas que cumplen con la condición de capitalización y posición.
print(f"Palabras encontradas (excluyendo la primera del comentario): {palabras_p10}")

# Mostramos el conteo total de coincidencias.
print(f"Total de resultados obtenidos: {len(palabras_p10)}")
```

*   **Pregunta 11.**  

Busca e imprime la secuencia de dos o más palabras que están separadas por un guion, "-", sin que tengan espacios en blanco entre ellas.

Por ejemplo "Go-Kart" sería válido, pero "Go  -Kart" o "Go  -  Kart" no lo serían. 

Indica la cantidad de resultados obtenidos.

```python
# Buscamos secuencias de dos o más palabras alfabéticas unidas por guion, sin espacios.

palabras_compuestas = re.findall(r'\b[A-Za-z]+(?:-[A-Za-z]+)+\b', " ".join(docs))
```

```python
# Imprimimos la lista de secuencias con guion encontradas para analizar cómo se estructuran estos términos en las reseñas.
print(f"Secuencias con guion encontradas: {palabras_compuestas}")

# Mostramos el total de resultados obtenidos.
print(f"Total: {len(palabras_compuestas)}")
```

*   **Pregunta 12.**  

Busca e imprime todas las palabras que terminan en "ing" o "ed". 

Indica la cantidad de palabras que encontraste de cada una.

```python
# Unimos el texto y ejecutamos ambas búsquedas en una sola línea usando asignación múltiple.
res_ing, res_ed = re.findall(r'\b\w+ing\b', " ".join(docs), re.I), re.findall(r'\b\w+ed\b', " ".join(docs), re.I)
```

```python
print(f"Palabras terminadas en 'ing': {res_ing}")
print(f"Total 'ing': {len(res_ing)}")

print(f"\nPalabras terminadas en 'ed': {res_ed}")
print(f"Total 'ed': {len(res_ed)}")
```

#**Parte 3. Proceso de limpieza.**

*   **Pregunta 13.**  

Ahora realiza un proceso de limpieza del corpus que incluya los siguientes procesos:

*   Solo se deben considerar caracteres alfabéticos. Es decir, se eliminan todos los signos de puntuación y caracteres especiales.
*   Todos los caracteres alfabéticos se transforman a minúsculas.
*   Se deben eliminar todos los espacios en blanco adicionales que se puedan encontrar en cada comentario.

Al finalizar dicho proceso de limpieza, imprime el resultado de los primeros 10 comentarios resultantes. 
   




```python
# Aplicamos re.sub para mantener solo letras y espacios, pasamos a minúsculas con .lower() 
# y normalizamos espacios en blanco usando .split() seguido de .join().
docs_clean = [" ".join(re.sub(r'[^A-Za-zÁÉÍÓÚÑáéíóúñ\s]', '', c).lower().split()) for c in docs]
```

```python
# Imprimimos los primeros 10 comentarios ya procesados para validar que la limpieza fue exitosa.
print(f"Muestra del corpus limpio:\n{docs_clean[:10]}")

# Mostramos el total de documentos para confirmar que mantenemos los 1000 registros originales.
print(f"Total de comentarios procesados: {len(docs_clean)}")
```

*   **Pregunta 14.**  

Con el resultado de la limpieza obtenido en la pregunta anterior, realiza ahora un proceso de tokenización por palabras del corpus. 

Es decir, al final de este proceso de tokenización, debes tener como resultado una lista de listas, donde cada comentario estará tokenizado por palabras.

Al terminar calcula el total de tokens obtenido en todo el corpus. 

```python
# Creamos una lista de listas donde cada sublista contiene las palabras (tokens) de un comentario.
# Usamos .split() que, por defecto, separa por cualquier espacio en blanco.
tokens = [c.split() for c in docs_clean]
```

```python
# Imprimimos los tokens de los primeros 5 comentarios para observar la estructura de lista de listas.
print(f"Muestra de la tokenización (primeros 5): {tokens[:5]}")

# Calculamos el total de tokens generados en todo el corpus para conocer el volumen de datos.
print(f"Total de tokens en el corpus: {sum(len(t) for t in tokens)}")
```

*   **Pregunta 15.**  

Finalmente, en este ejercicio definiremos nuestro conjunto de palabras "stopwords", las cuales deberás eliminar de todo el corpus. 

Recuerda que ejemplos de stopwords son artículos, adverbios, conectivos, etcétera, que tienen frecuencias de aparición muy altas en cualquier documento, pero que no brindan mucho significado en cuanto al significado de un enunciado.

Con base a la lista de stopwords que se te proporciona, realiza un proceso de limpieza eliminando todas estas palabras del corpus obtenido en el ejercicio anterior.

Obtener cuántos tokens/palabras quedan finalmente en todo el corpus.

Obtener cuántos de estos tokens/palabras son diferentes, es decir, cuántos tokens únicos tendrá lo que llamaremos más adelante nuestro vocabulario.

```python
# Considera la siguiente lista como tu conjunto de stopwords:
mis_stopwords = ['i', 'me', 'my', 'myself', 'we', 'our', 'ours', 'ourselves', 'you', 'your', 'yours', 'he', 'him', 'his', 'himself', 'she', 'her', 'hers', 'herself', 'it', 'its', 'itself', 'they', 'them', 'their', 'theirs', 'themselves', 'what', 'which', 'who', 'whom', 'this', 'that', 'these', 'those', 'am', 'is', 'are', 'was', 'were', 'be', 'been', 'being', 'have', 'has', 'had', 'having', 'do', 'does', 'did', 'doing', 'a', 'an', 'the', 'and', 'but', 'if', 'or', 'because', 'as', 'until', 'while', 'of', 'at', 'by', 'for', 'with', 'about', 'against', 'between', 'into', 'through', 'during', 'before', 'after', 'above', 'below', 'to', 'from', 'up', 'down', 'in', 'out', 'on', 'over', 'under', 'again', 'further', 'then', 'once', 'here', 'there', 'when', 'where', 'why', 'how', 'all', 'any', 'both', 'each', 'few', 'more', 'most', 'other', 'some', 'such', 'only', 'own', 'same', 'so', 'than', 'too', 'very', 's', 't', 'can', 'will', 'just', 'should', 'now', 'll']
```

```python
# Filtramos cada lista de tokens para mantener solo las palabras que NO están en nuestra lista de stopwords.
tokens_final = [[w for w in s if w not in mis_stopwords] for s in tokens]

# Generamos el vocabulario único final a partir de los tokens ya filtrados.
vocab_final = set(p for sub in tokens_final for p in sub)
```

```python
# Imprimimos el gran total de palabras que quedaron en todo el corpus tras la limpieza.
print(f"Tokens totales finales: {sum(len(t) for t in tokens_final)}")

# Mostramos el tamaño del vocabulario final.
print(f"Tamaño del vocabulario único: {len(vocab_final)}")
```
